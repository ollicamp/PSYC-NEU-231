1) From the "tips" dataset
    1. add a percentage tip column to the dataset
    2. visualize the relationship between tip percentage, with day of week and time 
    3. visualize tip ~ day, time, and number of people (which function allows three variables independent variables at once?)


```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
tips = sns.load_dataset('tips')
print(tips.shape)
tips.head()
```

```python
ptip = tips.tip.values/tips.total_bill.values
print(ptip.shape)

tips['per_tip'] = ptip
tips.shape
tips.head()
```

```python
# seems like the only meals are lunch and dinner
# only days are Thurs-Sun?
sns.pointplot(x='day',y='per_tip',hue='time',data=tips)
plt.show()
```

```python
sns.catplot(x='size',y='per_tip',hue='day',col='time',data=tips,kind='point')
```

```python
# for dinner it seems like percent tip decreased with group size
# wonder if that has to do with the total bill amount
sns.jointplot(x='total_bill',y='per_tip',data=tips) 
plt.show()
# huh it looks like that's being driven by a few really high tippers, probably the individual diners
```

```python
sns.catplot(x='day',y='per_tip',hue='size',col='time',data=tips,kind='point')
```

2) Find an interesting trend in the "exercise" dataset

```python
exercise = sns.load_dataset('exercise')
print(exercise.shape)
exercise.head()
# so I guess they collected heart rates after 1, 15, or 30 min of rest, walking, or running
# everyone is either on low fat or no fat diet
```

```python
sns.catplot(x='kind',y='pulse',hue='time',col='diet',data=exercise,kind='point')
# seems pretty expected in terms of activity (heart rate increases with activity level and amount)
# the increase seems much smaller for the low fat diet than no fat
```

```python
# try to pull that out better ie overlay diets
sns.catplot(x='time',y='pulse',hue='diet',col='kind',data=exercise,kind='point')
plt.show()
# no fat consistently has higher heart rate
# effect is more pronounced when running, separates more with more running
```
