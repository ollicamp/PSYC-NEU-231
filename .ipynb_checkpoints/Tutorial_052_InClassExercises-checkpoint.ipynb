## Are there missing values (NaNs) in the data?

```python
if np.isnan(respt).any:
    print('TRUE')
# okay so there are some...
    
nans = pd.isna(respt)
np.sum(nans, axis=0)
# and that's how many for each subject
```

## What about outliers? Lets define outliers here as > 2 * std away from the mean for each subject
* After you've found the outliers for each subject, replace those values with a np.nan (NaN)

```python
# try it on one subject first
this_dat = respt['Sub0']
submean = np.mean(this_dat)
substd = np.std(this_dat)
print('Initial NaNs:', sum(pd.isna(this_dat)))
# no missing data for this subject, which we knew

for i in np.arange(len(this_dat)):
    outliers = (np.abs(this_dat[i] - submean) > 2*substd)
    this_dat[outliers] = np.nan
print('Outlier NaNs: ', sum(pd.isna(this_dat)))
# replaced two values with NaN
```

```python
# now to do it for each subject
# loop over each column
# use respt.columns to grab column names
# need to remember to replace values in respt not the temporary vector
file_name = cwd + '/response_time_data.csv'
respt = pd.read_csv(file_name, index_col=0, header=0)

for col in respt.columns:
    # work on respt[col]
    submean = np.mean(respt[col])
    substd = np.std(respt[col])
    for i in np.arange(len(respt[col])):
        outliers = (np.abs(respt[col][i] - submean) > 2*substd)
        respt[col][outliers] = np.nan
    print(sum(pd.isna(respt[col])))
    # so we can see how many outliers there are for each subject
```

## After you've found the outliers and replaced with NaNs for each subject, check out this function:
[pandas.DataFrame.interpolate](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.interpolate.html#pandas.DataFrame.interpolate)

* Use this function to interpolate the missing values for each subject (do not interpolate across subjects!)
* Just use linear interpolation...

```python
# interpolate across columns (within a subject)
respt2 = respt.interpolate(method='linear', axis=1)
summary2 = respt2.describe()
display(summary2)
```

## You can explore the "Missing Values" page for Pandas to figure out other ways of filling in missing values and outliers

[page is here](https://pandas.pydata.org/pandas-docs/stable/missing_data.html#missing-data)

* Figure out how to replace the outliers with the mean of each subject

```python
# replace the missing values with subject mean
for col in respt.columns:
    respt[col].fillna(np.mean(respt[col]))
```

## Use the Pandas.DataFrame.Sample function to generate bootstrapped confidence intervals for the data from subject 11

[see this page for Samples](https://pandas.pydata.org/pandas-docs/version/0.22/generated/pandas.DataFrame.sample.html)


* Resample Sub11's data with replacement, each time pulling N samples (800 in this case)
* Generate a distribution of means across all resamples
* Compute 95% confidence intervals using:

[this page for quantiles](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.quantile.html)

```python
nboot = 10000
N = 800
tempbt = np.zeros(N)
btmeans = np.zeros(nboot)

for i in np.arange(nboot):
    tempbt = respt['Sub11'].sample(n=800, replace=True)
    btmeans[i] = np.mean(tempbt)
    
btCI = np.percentile(btmeans, [2.5, 97.5])
```

```python
print('Subject mean: ', np.mean(respt['Sub11']))
print('Confidence Interval: ', btCI[0], btCI[1])

plt.hist(btmeans, bins=30)
plt.axvline(np.mean(respt['Sub11']), color='k', linewidth=1)
plt.axvline(btCI[0], color='r', linewidth=1, linestyle='--')
plt.axvline(btCI[1], color='r', linewidth=1, linestyle='--')
plt.show()
```
